```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import rcParams
%matplotlib inline
config = {
    "font.family":'serif',
    "font.size": 14,
    "mathtext.fontset":'stix',
    "font.serif": ['SimSun'],
}
rcParams.update(config)

legend_english_font = {"family" : "Times New Roman"}
```

# Accuracy

```python
# 数据
count = [i+1 for i in range(5)]
bayes_Accuracy = [0.858, 0.858, 0.858, 0.858, 0.858]
svm_Accuracy = [0.856, 0.856, 0.856, 0.856, 0.856]
xgboost_Accuracy = [0.846, 0.846, 0.846, 0.846, 0.846]
lstm_Accuracy = [0.878, 0.854, 0.87, 0.852, 0.872]
bert_Accuracy = [0.864, 0.862, 0.866, 0.864, 0.862]
bert_lstm_Accuracy = [0.884, 0.874, 0.868, 0.88, 0.87]
```

```python
# 画图
plt.figure(figsize=(7.5, 5.5))
plt.plot(count, bayes_Accuracy, '#f368e0',marker='*', markersize=5)
plt.plot(count, svm_Accuracy, '#ff9f43',marker='*', markersize=5)
plt.plot(count, xgboost_Accuracy, '#222f3e',marker='*', markersize=5)
plt.plot(count, lstm_Accuracy, '#0abde3',marker='*', markersize=5)
plt.plot(count, bert_Accuracy, '#10ac84',marker='*', markersize=5)
plt.plot(count, bert_lstm_Accuracy, '#ee5253',marker='*', markersize=5)

plt.title('各模型准确率（Accuracy）五次结果')  # 折线图标题
plt.xlabel('次数')  # x轴标题
plt.ylabel('准确率')  # y轴标题

#给图像添加注释，并设置样式
for a, b in zip(count, bayes_Accuracy):
    plt.text(a, b, b, ha='center', va='bottom', fontsize=10)
for a, b in zip(count, svm_Accuracy):
    plt.text(a, b, b, ha='center', va='bottom', fontsize=10)
for a, b in zip(count, xgboost_Accuracy):
    plt.text(a, b, b, ha='center', va='bottom', fontsize=10)
for a, b in zip(count, lstm_Accuracy):
    plt.text(a, b, b, ha='center', va='bottom', fontsize=10)
for a, b in zip(count, bert_Accuracy):
    plt.text(a, b, b, ha='center', va='bottom', fontsize=10)
for a, b in zip(count, bert_lstm_Accuracy):
    plt.text(a, b, b, ha='center', va='bottom', fontsize=10)

# 绘制图例
plt.legend(['bayes', 'svm', 'xgboost', 'lstm', 'bert', 'bert_lstm'], prop=legend_english_font)
# 保存图像
plt.savefig('./visualization/title/Accuracy.jpg')
# 显示图像
plt.show()
```
